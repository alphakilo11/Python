```
from matplotlib import pyplot

with open("test.txt") as file:
  foo = file.readlines()
bar = []
for i in foo:
  bar.append(i.split("\t"))
eggs = []
for i in bar:
  eggs.append(float(i[1]))

#print(foo)
#print(bar)
#print(eggs)
steps = []
for i in range(1, len(eggs)):
  steps.append(eggs[i - 1] - eggs[i])
print (steps)
x = range(len(eggs))
pyplot.plot(x, eggs)
pyplot.title("Usain Bolt's 200-m progression")
```

```
import numpy as np
import matplotlib.pyplot as plt

with open("jan.txt") as file:
  foo = file.readlines()
bar = []
for i in foo:
  bar.append(i.split("\t"))
y = []
for i in bar:
  y.append(float(i[1]))

#print(foo)
#print(bar)
#print(eggs)

x = range(len(y))
plt.scatter(x, y)
plt.title("Jan Zelezny's Javelin throw progression")
#calculate equation for trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#add trendline to plot
plt.plot(x, p(x))
```

```
import numpy as np
import matplotlib.pyplot as plt

#define data
x = np.array([8, 13, 14, 15, 15, 20, 25, 30, 38, 40])
y = np.array([5, 4, 18, 14, 20, 24, 28, 33, 30, 37])

#create scatterplot
plt.scatter(x, y)

#calculate equation for trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#add trendline to plot
plt.plot(x, p(x))
```

```
speed_100m = 100 / 9.58
levels = 100
increment = speed_100m / levels
y = []
for i in range(1, levels + 1):
  y.append(i * increment)
x = range(len(y))
plt.plot(x, y)
```
